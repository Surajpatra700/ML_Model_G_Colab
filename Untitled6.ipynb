```python
from sklearn import svm, datasets
iris = datasets.load_iris()
import pandas as pd
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['flower'] = iris.target
df['flower'] = df['flower'].apply(lambda x: iris.target_names[x])
df[47:52]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.3)
```

```python
model = svm.SVC(kernel='rbf',C=10,gamma='auto')
model.fit(X_train,y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(svm.SVC(kernel='linear',C=10, gamma='auto'),iris.data, iris.target,cv=5) # cv ---> cross validation/ no. of iterations
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=10, gamma='auto'),iris.data, iris.target,cv=5)
```

```python
import numpy as np
kernels = ['rbf','linear']
C = [1,10,20]
avg_scores = {}
for kval in kernels:
  for cval in C:
    cv_scores = cross_val_score(svm.SVC(kernel=kval,C = cval, gamma='auto'),iris.data, iris.target, cv=5)
    avg_scores[kval +"_"+ str(cval)] = np.average(cv_scores)
avg_scores
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma='auto'),{
    'C': [1,10,20],
    'kernel': ['rbf','linear']
}, cv=5)

clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_C',"param_kernel","mean_test_score"]]
```

```python
dir(clf)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
# GridSearchCV tries for all possible combination of kernel and C, hence computation is more, less efficient to overcome that we use RandomSearchCV
# It randomly takes the value of kernel and C and show o/p based upon n_iter value
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(svm.SVC(gamma='auto'),{
    'C': [1,10,20],
    'kernel': ['rbf','linear']
}, cv=5, n_iter=2)

rs.fit(iris.data,iris.target)
pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params':{
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params':{
            'n_estimators': [1,5,10]
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C': [1,5,10]
        }
    }
}
```

```python
scores = []

for model_name, mp in model_params.items():
  clf = GridSearchCV(mp['model'],mp['params'],cv=5)
  clf.fit(iris.data, iris.target)
  scores.append({
      'model': model_name,
      'best_score': clf.best_score_,
      'best_params': clf.best_params_
  })
```

```python
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# read Dataset
dataset = pd.read_csv('/content/drive/MyDrive/datasets/melb_data.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
dataset = dataset[cols_to_use]
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset.head()
```

```python
X = dataset.drop('Price',axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X,train_y)
reg.score(test_X,test_y)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50,max_iter=100,tol=0.1) # Lasso regression refer to L1 Regularization
lasso_reg.fit(train_X,train_y)
```

```python
lasso_reg.score(test_X,test_y)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha =50, max_iter=100,tol=0.1)
ridge_reg.fit(train_X, train_y)
```

```python
ridge_reg.score(test_X, test_y)
```

```python
# PCA (Principal Component Analysis)

# Is a process of figuring out most important features or principal components that has the most impact on the target variable
# PCA is a dimensionality reduction technique as it can help us reduce dimensions.
```

```python
import pandas as pd
from sklearn.datasets import load_digits
dataset = load_digits()
dataset.data[0].reshape(8,8)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.gray()
plt.matshow(dataset.data[0].reshape(8,8))
```

```python
import numpy as np

np.unique(dataset.target)
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
df.describe()
```

```python
X = df
y = dataset.target
y
```

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler scales the value from -1 to 1 where as MinMaxScaler Scales from 0 to 1

scaler = StandardScaler()
X_Scaled = scaler.fit_transform(X)
X_Scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_Scaled, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test,y_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)   # PCA(0.95) extracts 95% essential data containing columns from all 64 columns
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca,y,test_size=0.2, random_state=30)
```

```python
model = LogisticRegression(max_iter = 1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```
