```python
from sklearn import svm, datasets
iris = datasets.load_iris()
import pandas as pd
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['flower'] = iris.target
df['flower'] = df['flower'].apply(lambda x: iris.target_names[x])
df[47:52]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.3)
```

```python
model = svm.SVC(kernel='rbf',C=10,gamma='auto')
model.fit(X_train,y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(svm.SVC(kernel='linear',C=10, gamma='auto'),iris.data, iris.target,cv=5) # cv ---> cross validation/ no. of iterations
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=10, gamma='auto'),iris.data, iris.target,cv=5)
```

```python
import numpy as np
kernels = ['rbf','linear']
C = [1,10,20]
avg_scores = {}
for kval in kernels:
  for cval in C:
    cv_scores = cross_val_score(svm.SVC(kernel=kval,C = cval, gamma='auto'),iris.data, iris.target, cv=5)
    avg_scores[kval +"_"+ str(cval)] = np.average(cv_scores)
avg_scores
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma='auto'),{
    'C': [1,10,20],
    'kernel': ['rbf','linear']
}, cv=5)

clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_C',"param_kernel","mean_test_score"]]
```

```python
dir(clf)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
# GridSearchCV tries for all possible combination of kernel and C, hence computation is more, less efficient to overcome that we use RandomSearchCV
# It randomly takes the value of kernel and C and show o/p based upon n_iter value
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(svm.SVC(gamma='auto'),{
    'C': [1,10,20],
    'kernel': ['rbf','linear']
}, cv=5, n_iter=2)

rs.fit(iris.data,iris.target)
pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params':{
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params':{
            'n_estimators': [1,5,10]
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C': [1,5,10]
        }
    }
}
```

```python
scores = []

for model_name, mp in model_params.items():
  clf = GridSearchCV(mp['model'],mp['params'],cv=5)
  clf.fit(iris.data, iris.target)
  scores.append({
      'model': model_name,
      'best_score': clf.best_score_,
      'best_params': clf.best_params_
  })
```

```python
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

```python
# read Dataset
dataset = pd.read_csv('/content/drive/MyDrive/datasets/melb_data.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
dataset = dataset[cols_to_use]
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset.head()
```

```python
X = dataset.drop('Price',axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X,train_y)
reg.score(test_X,test_y)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50,max_iter=100,tol=0.1) # Lasso regression refer to L1 Regularization
lasso_reg.fit(train_X,train_y)
```

```python
lasso_reg.score(test_X,test_y)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha =50, max_iter=100,tol=0.1)
ridge_reg.fit(train_X, train_y)
```

```python
ridge_reg.score(test_X, test_y)
```

```python
# ***************************** BAGGING AND BOOTING *******************

# Bagging, short for Bootstrap Aggregating, is a machine learning ensemble technique that aims to improve the accuracy and robustness of models by training multiple instances of the same algorithm on different subsets of the training data. It involves:

# 1. Bootstrapping: Creating multiple random samples (with replacement) from the original training dataset.
# 2. Training: Training a separate model on each of these bootstrap samples.
# 3. Aggregating: Combining the predictions of these models (e.g., by averaging for regression or voting for classification) to make a final prediction.

# Bagging helps reduce overfitting and improve model stability, making it a valuable technique for improving the performance of various machine learning algorithms, such as decision trees, in the ensemble. One popular algorithm that uses bagging is the Random Forest.
```

```python
X = df.drop('Outcome', axis = 'columns')
y = df.Outcome

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Printing 1st 3 rows out of it
X_scaled[:3]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, random_state=10)
```

```python
X_train.shape
```

```python
y_train.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

scores = cross_val_score(DecisionTreeClassifier(), X, y, cv=5)
scores
```

```python
# Bagging Algo

from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    base_estimator = DecisionTreeClassifier(),
    n_estimators = 100,
    max_samples = 0.8,
    oob_score =True,
    random_state = 0
)

bag_model.fit(X_train, y_train)
bag_model.oob_score_
```
