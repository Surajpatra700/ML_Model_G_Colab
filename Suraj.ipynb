```python
# Gradient Descent is the technique to find the best fit line of any i/p and o/p expression i.e slope and intercept / realational expression

import numpy as np

def gradient_decscent(x,y):
  # Initialising current value of m and b as 0
  m_curr = b_curr = 0
  iterations = 1000
  n = len(x) # len(x) = len(y)
  # Lets consider learning rate initial value as 0.001, generally we keep learning rate as less as possible
  learning_rate = 0.001
  for i in range(iterations):
    y_predicted = m_curr * x + b_curr
    cost = 1/n * sum([val**2 for val in (y-y_predicted)])
    # md --> m derivation i.e d/dm & bd --> b derivation i.e d/db
    md = -(2/n)*sum(x*(y - y_predicted))
    bd = -(2/n)*sum(y - y_predicted)
    m_curr =  m_curr - learning_rate * md
    b_curr = b_curr - learning_rate*bd
    print("m {}, b {}, cost{}, iteration {}".format(m_curr, b_curr,cost,i))

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_decscent(x,y)
```

```python
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np

digits = load_digits()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.3)
lr = LogisticRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
svm = SVC()
svm.fit(X_train,y_train)
svm.score(X_test,y_test)
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rf.score(X_test,y_test)

def get_score(model, X_train,X_test,y_train,y_test):
  model.fit(X_train,y_train)
  return model.score(X_test,y_test)

get_score(LogisticRegression(),X_train,X_test,y_train,y_test)

# cross_val_score
# Cross value algorithm tell us the performance of multiple algos to anyt given dataset
from sklearn.model_selection import cross_val_score
cross_val_score(LogisticRegression(), digits.data, digits.target)
cross_val_score(SVC(), digits.data, digits.target)
# RandomForestClassifier ran with 40 trees
# giving n_estimators is called parameter tuning
cross_val_score(RandomForestClassifier(n_estimators=40), digits.data, digits.target)
```

```python
# *************************** K Means (UnSupervised Learning) ************************

from pandas.core.strings.accessor import Scalar
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
df = pd.read_csv("/content/drive/MyDrive/datasets/k_means.csv")
df
# plt.scatter(df['Age'],df['Income($)'])
# KMeans convert the dataset into n_Clusters = 3.... i.e 0 to 3
# km = KMeans(n_clusters=3)
# km
# y_predicted = km.fit_predict(df[['Age','Income($)']])
# y_predicted
# df['cluster'] = y_predicted
# df

# df0 = df[df.cluster==0]
# df1 = df[df.cluster==1]
# df2 = df[df.cluster==2]

# plt.scatter(df0.Age, df0['Income($)'],color='green')
# plt.scatter(df1.Age, df1['Income($)'],color='red')
# plt.scatter(df2.Age, df2['Income($)'],color='orange')
# plt.xlabel('Age')
# plt.ylabel('Income($)')
# plt.legend()

# MinMaxScaler scales the dataset labels to range of 0 to 1
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
df

scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
df

km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
df['cluster'] = y_predicted
df

df0 = df[df.cluster==0]
df1 = df[df.cluster==1]
df2 = df[df.cluster==2]

# Centroids of 3 Clusters
km.cluster_centers_

plt.scatter(df0.Age, df0['Income($)'],color='green',label='Income($)')
plt.scatter(df1.Age, df1['Income($)'],color='red',label='Income($)')
plt.scatter(df2.Age, df2['Income($)'],color='orange',label='Income($)')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()

k_rng = range(1,10)
# sum of square of errors(sse)
sse = []
for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[['Age','Income($)']])
  # km.inertia_ gives the sum of square of errors(sse)
  sse.append(km.inertia_)


plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

```python
# Naive Bayes for Spam Detection

import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/datasets/spam.csv")
df.head()
# Categorizing the data into spam and ham

df.groupby('Category').describe()
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.Message,df.spam,test_size=0.2)

# Converting Message text to numbers using COUNT VECTORIZER TECHNIQUE
# It converts all the unique words in message column to features(columns) containing column matrix/vector

from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:3]

from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)

email = [
    'Hey mohan, can we get together to watch Football game tommorow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
# Converting the ermail/message column to document-term matrix using transform method
email_count = v.transform(email)
model.predict(email_count)

# Predicting the accuracy of model, converting email/messages into feature matrix

X_test_count = v.transform(X_test)
model.score(X_test_count,y_test)
```
